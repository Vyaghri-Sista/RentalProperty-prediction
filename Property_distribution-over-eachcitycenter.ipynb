```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
path1=os.getcwd()

#Loading the data
path=path1+'/file/Train.csv'
data=pd.read_csv(path)


#data['neighbourhood_group'].unique()
#dpr_mn1=data[(data['room_type']=='Private room')&(data['neighbourhood_group']=='Manhattan')]
#dpr_mn2=data[(data['room_type']=='Entire home/apt')&(data['neighbourhood_group']=='Manhattan')]
#dpr_mn3=data[(data['room_type']=='Shared room')&(data['neighbourhood_group']=='Manhattan')]
```

```python
dat_m=data[data['neighbourhood_group']=='Manhattan']
dat_B=data[data['neighbourhood_group']=='Brooklyn']
dat_Q=data[data['neighbourhood_group']=='Queens']
dat_Si=data[data['neighbourhood_group']=='Staten Island']
dat_Bn=data[data['neighbourhood_group']=='Bronx']
```

# calculation of distance between two lat and long

```python
dr2=dat_m[['latitude','longitude']]
dr3=dr2.to_numpy()
```

# Distance calculation from the city center (Manhattan City Center: 40.7831,-73.9712).

```python
import math
dr5=dr3*math.pi/180

dr4=dr3*math.pi/180
#dr4
dr4[:,0] -= 40.7831*math.pi/180
dlat=dr4[:,0]
dlon=dr4[:,1]

dlon -= -73.9712*math.pi/180

a=[]
for i in range(len(dr3)):
    a.append(math.sin(dlat[i]/2)**2+math.cos(dr5[i][0])*math.cos(40.7831*math.pi/180)*math.sin(dlon[i]/2)**2)
    
c=[]
for i in range(len(dr3)):
    c.append(math.atan2(math.sqrt(a[i]),math.sqrt(1-a[i])))
    
d=[]
for i in range(len(dr3)):
    d.append(6373.0*c[i])
```

# Adding the distance numpy array to the Manhattan dataframe

```python
dat_m['Distance in km']= d
```

# Plotting the properties from the city center of Manhattan- private room

```python
dpr_mn1=dat_m[dat_m['room_type']=='Private room']
figure =plt.figure(figsize=(15,8))
sns.distplot(dpr_mn1['Distance in km'],bins=60)
plt.title('Distribution of Properties as a function of distance from City Center', fontsize=30)
plt.xlabel('Distance in (km)',fontsize=25)
plt.ylabel('Distribution',fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

```python
dpr_mn2=dat_m[dat_m['room_type']=='Entire home/apt']
figure =plt.figure(figsize=(15,8))
sns.distplot(dpr_mn2['Distance in km'],bins=60)
plt.title('Distribution of Properties as a function of distance from City Center', fontsize=30)
plt.xlabel('Distance in (km)',fontsize=25)
plt.ylabel('Distribution',fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

```python
dpr_mn3=dat_m[dat_m['room_type']=='Shared room']
figure =plt.figure(figsize=(15,8))
sns.distplot(dpr_mn3['Distance in km'],bins=60)
plt.title('Distribution of Properties as a function of distance from City Center', fontsize=30)
plt.xlabel('Distance in (km)',fontsize=25)
plt.ylabel('Distribution',fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```
